```python
## Fix data

from datasets import load_dataset
import pandas as pd
version = 5
file = f'L3_data{version}.csv'
data  = load_dataset("csv", data_files = file) #Load dataset
data = pd.DataFrame(data['train'])
data =data.drop(columns = "Unnamed: 0")
display(data)
```

```python
## Replace values

## Find N
N = data.shape[0]

for i in range(N):
    val = data.loc[i, "Confidence"]
    
    val = val / 100
    data.loc[i, "Confidence"] = val



display(data)
```

```python
new_file_name = f'L3_data{version}.csv'
data.to_csv(new_file_name)
```
